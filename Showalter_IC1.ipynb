# Establish the Database Connection

```python
#connect to SQLite3
import sqlite3
db_path = '/content/drive/MyDrive/ScoutingProject/scouting_data.db'
conn = sqlite3.connect(db_path)
c = conn.cursor()
print("Permanent database connection established")
```

# Create the Table Structure to create the heading.

```python
#define the schema
c.execute(''' CREATE TABLE IF NOT EXISTS quarterbacks (
     name TEXT,
     team TEXT,
     passing_yards INTEGER,
     touchdowns INTEGER
     )
     ''')
  #Save by commiting
conn.commit()
print("table Quarterbacks is ready")
```

# Populate Data and Save

```python
qb_list = ('Josh Allen', 'Buffalo Bills', 4306, 29)
c.execute('INSERT INTO quarterbacks VALUES (?, ?, ?, ?)', qb_list)
conn.commit()
print("Successfully added qb Josh Allen")
```

# Run the Scouting Report

```python
#Querying the data
c.execute('SELECT * FROM quarterbacks')
results = c.fetchall()
for row in results:
  print(row)
```

# Proper Shutdown - Important!

```python
conn.close()
print("Database safely closed. You can now see this .db file in your Google Drive folder.")
```
